```python
# Reading in test.csv and saving as test.xlsx

class CSVtoXLS :
    
    def __init__(self, filename, sep=";", encoding="utf-8") : 
        self.filename = filename
        self.file = filename + ".csv"
        self.sep = sep
        self.encoding = encoding
        self.df = None
        self.columnsUniques = []
        
    def toDF(self): 
        self.df = pd.read_csv(self.file, sep=self.sep, encoding=self.encoding)
        return df
    
    def getUniques(self):
        results=[]
        for col in self.df:
            uniques = self.df[col].unique()
            u_dict = {"colname" : col, "uniques" : uniques.tolist() }
            results.append(u_dict)
        self.columnsUniques = results
        return results
            
    def toXLS(self):
        writer = pd.ExcelWriter(self.filename + '.xlsx')
        self.df.to_excel(writer, index = False)
        writer.save()
```

```python
CSVlist = [CSVfiles[0]]

CSV_dfList = []

for csv in CSVlist:
    classCSV = CSVtoXLS(csv["name"],sep=csv["sep"],encoding=csv["encoding"] )
    classCSV.toDF()
    classCSV.getUniques()
    result = { 
        "code" : csv["code"], 
        "df" : classCSV.df, 
        "columnsUniques" : classCSV.columnsUniques,
        "classCSV" : classCSV, 
        "csvInfos" : csv
    }
    CSV_dfList.append(result)
```
